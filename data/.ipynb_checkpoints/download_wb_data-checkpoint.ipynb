```python
from IPython.display import display, HTML
from datetime import date
import os
import pandas as pd

# pandas_datareader spams FutureWarning on import
import warnings
warnings.filterwarnings('ignore', module='pandas_datareader')
from pandas_datareader import wb

# find target path for data files, assuming the notebook is in the right place
data_path = _dh[0]
assert data_path.endswith(os.path.join('em-2020','data'))
```

Fetch all indicators and select out those of interest:

```python
all_indicators = wb.get_indicators().set_index('id')
# all_indicators = all_indicators.query('source == "World Development Indicators"')
# all_indicators = all_indicators.query('source == "WDI Database Archives"')
indicator_map = {
#     'NV.IND.TOTL.ZS': 'ISG',
#     'EG.EGY.PRIM.PP.KD': 'EI',
#     '6.1_PRIMARY.ENERGY.INTENSITY': 'EI',
    'NE.TRD.GNFS.ZS': 'TO',
    'NY.GDP.PCAP.KD': 'P_GDPa',
    'NY.GDP.PCAP.CD': 'P_GDP',
}
indicator_map = {}
search = lambda x: all_indicators.loc[all_indicators.name.str.match(x, case=False), 'name']
to_test = {k:k for k in search('indust').index}
indicator_map.update(to_test)
indicator_ids = indicator_map.keys()
all_indicators.loc[indicator_ids, ['name', 'sourceNote', 'source']]
```

Fetch all countries and select out those of interest:

```python
all_countries = wb.get_countries().set_index('iso3c')
country_ids = [
    'BGD',
#     'GBR',
]
country_map = all_countries.loc[country_ids, 'name'].to_dict()
all_countries.loc[country_ids, ['name', 'region']]
```

Download the data and save it 

```python
data = wb.download(country=country_ids, indicator=indicator_ids, start=1986, end=date.today().year)
```

```python
# rescale to 2018 US$ levels to match Pan et al. (2019)
scaler = 1210.16 / 1248.453398
data['NY.GDP.PCAP.CD'] *= scaler

data.rename(columns=indicator_map, inplace=True)

for country_id, country_name in country_map.items():
    country_data = data.xs(country_name, level='country').sort_index().dropna(how='all')
    country_data.index = country_data.index.astype('i')
    display(HTML('<h2>{}</h2>'.format(country_name)))
    display(country_data.head())
    display(HTML(country_data.loc[1986:2015].describe().loc[['min', 'max', 'mean', 'std']].T
            .sort_values('min').to_html(float_format='{:.2f}'.format)))
    country_data.to_csv(os.path.join(data_path, '{}.csv'.format(country_name)))
```
